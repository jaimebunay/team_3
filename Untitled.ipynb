```python
import pandas as pd 
from pathlib import Path 
import numpy as np
```

```python
state_unemployment = pd.read_csv(Path('Resources/cleaned_unemployment_data.csv'), index_col = 'State')
state_unemployment.tail()
```

```python
state_unemployment.drop('District of Columbia', inplace = True)
```

```python
#s_result = d.DataFrame({"avg": s_rev.mean(axis=1), "sd": s_rev.std(axis=1)})
results = pd.DataFrame({'avg':state_unemployment.pct_change(axis=1).mean(axis=1), 'std':state_unemployment.pct_change(axis=1).std(axis=1)})
```

```python
results.head()
```

```python
avg = results['avg']
```

```python
std = results['std']
```

```python
monte_carlo = pd.DataFrame()
```

```python
'''DOING TEN SIMULATIONS FOR EACH STATE'''
num_simulation = 25
num_years = 10

x = 0
for x in range(50):
    # Getting the last unemployment from each state
    index = state_unemployment.index[x]
    last_rate = state_unemployment.loc[index, '2019']
    
    #Running monte carlo simulation
    for num in range(num_simulation): 
        simulated_rate = [last_rate]
        #spy_randoms = np.random.normal(avg_daily_return_spy, volatility_return_spy, number_records)
        #agg_randoms = np.random.normal(avg_daily_return_agg, volatility_return_agg, number_records)
        #randon = np.random.normal(avg[x],std[x], num_simulation) 
        for year in range(num_years): 
            lista = []
            for i in range(num_simulation): 
                rate = simulated_rate[-1]*(1 + np.random.normal(avg[x], std[x]))
                lista.append(rate)
                           
            simulated_rate.append(sum(lista)/len(lista))
        #lista.append(simulated_rate)
    
    monte_carlo[index] = pd.Series(simulated_rate)
    x =+ 1
```
