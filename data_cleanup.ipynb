```python
from pathlib import Path 
import pandas as pd
%matplotlib inline
import os
import plotly.express as px
```

```python
# Import Unemployment Data from Resources, without the header
state_unemployment = pd.read_csv(Path('Resources/state_unemployment_rate_data.csv'), header = None)
state_unemployment.drop(0, inplace = True)
#state_unemployment.set_index(0)
state_unemployment.head(5)
```

```python
# Flip the axis of the the DataFrame, States names will go to columns
state_unemployment = state_unemployment.transpose()
```

```python
state_unemployment.head()
```

```python
# As seen above, State names are in row 0, so in order to grab them we use loc. 
column=state_unemployment.loc[0]
```

```python
# set column names
state_unemployment.columns =column
state_unemployment.head()
```

```python
# Drop row 0 now
state_unemployment.drop(0, inplace = True)
```

```python
# Drop row one because this row contains data from Dec 2009, which won't be part of my dataframe
state_unemployment.drop(1,inplace = True)
```

```python
state_unemployment.reset_index(inplace =True)
state_unemployment.head()
```

```python
# drop the index column
state_unemployment.drop(columns='index', inplace = True)
state_unemployment.head()
```

```python
# For the purpose of our project, we don't consider Puerto Rico 
state_unemployment.drop(columns = 'Puerto Rico', inplace = True)
```

```python
# Lets see what kind of data is in our dataframe
state_unemployment.dtypes
```

```python
# Change the entire dataframe into float format
state_unemployment= state_unemployment.astype(float)
```

```python
# lets create a useful index for our dataframe.
index = pd.date_range(start='1/1/2010', end='12/31/2019', freq='M' )
index
```

```python
# set our new index
state_unemployment.set_index(index, inplace = True)
state_unemployment.head()
```

```python
state_unemployment.shape
```

```python
""" 
The original csv file contained montly uemployment rates for the last ten years.
We decided that grouping it by year was going to be more meaninful for the purpose of our project.
"""
# use groupby function to group by year and calcalute the avg 
# unemployment rate for each year
state_unemployment = state_unemployment.groupby(state_unemployment.index.year).mean()
```

```python
cleaned_data = state_unemployment.round(decimals=2)
cleaned_data
```

```python
# Label our index
cleaned_data.index.rename('Year', inplace = True)
```

```python
cleaned_data.head()
```

```python
cleaned_data=cleaned_data.transpose()
```

```python
cleaned_data.head()
```

```python
# give our index a name
cleaned_data.index.name = 'State'
```

```python
cleaned_data.head()
```

```python
# Save cleaned data as a new csv file
cleaned_data.to_csv("Resources/cleaned_unemployment_data.csv")
```
